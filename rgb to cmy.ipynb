```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image_path=r"C:\Users\student\Pictures\cv\pic 1.webp"
image=Image.open(image_path)
rgb_arr=np.array(image) 
cmy_arr=255-rgb_arr
cmy_image=Image.fromarray(cmy_arr.astype('uint8'),'RGB')
cmy_path=r"C:\Users\student\Pictures\cv\cmy_image1.webp"
cmy_image.save(cmy_path)
plt.figure(figsize=(5,7))
r_channel, g_channel, b_channel = image.split()

r_arr = np.array(r_channel)
g_arr = np.array(g_channel)
b_arr = np.array(b_channel)

c_channel = 255 - r_arr  
m_channel = 255 - g_arr  
y_channel = 255 - b_arr  

c_image=Image.fromarray(c_channel.astype('uint8'),'L')
m_image=Image.fromarray(m_channel.astype('uint8'),'L')
y_image=Image.fromarray(y_channel.astype('uint8'),'L')
c_path=r"C:\Users\student\Pictures\cv\c_channel.png"
m_path=r"C:\Users\student\Pictures\cv\m_channel.png"
y_path=r"C:\Users\student\Pictures\cv\y_channel.png"
c_image.save(c_path)
m_image.save(m_path)
y_image.save(y_path)
plt.figure(figsize=(12, 4))

plt.subplot(1,4,1)
plt.imshow(cmy_image)
plt.title('CMY image')
        
plt.subplot(1, 4, 2)
plt.imshow(c_image)
plt.title('Cyan Channel') 

plt.subplot(1, 4, 3)
plt.imshow(m_image)
plt.title('Magenta Channel') 

plt.subplot(1, 4, 4)
plt.imshow(y_image)
plt.title('Yellow Channel')

plt.show()
```
