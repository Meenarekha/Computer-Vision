```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread(r'/content/my image.jpg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

hsvimg = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
hue= 10
saturation = 50
value = 30
h, s, v = cv2.split(hsvimg)
h = cv2.add(h, hue)
s = cv2.add(s, saturation)
v = cv2.add(v, value)
img_hsv_modified = cv2.merge((h, s, v))
img_rgb_modified = cv2.cvtColor(img_hsv_modified, cv2.COLOR_HSV2RGB)
plt.figure(figsize=(15, 15))


plt.subplot(3, 3, 1)
plt.imshow(img)
plt.title('Original RGB Image')
plt.axis('off')


plt.subplot(3, 3, 2)
plt.imshow(h, cmap='gray')
plt.title('Hue Channel')
plt.axis('off')


plt.subplot(3, 3, 3)
plt.imshow(s, cmap='gray')
plt.title('Saturation Channel')
plt.axis('off')


plt.subplot(3, 3, 4)
plt.imshow(v, cmap='gray')
plt.title('Value Channel')
plt.axis('off')


plt.subplot(3, 3, 5)
plt.imshow(img_hsv_modified)
plt.title('Modified HSV Image')
plt.axis('off')


plt.subplot(3, 3, 6)
plt.imshow(img_rgb_modified)
plt.title('Modified RGB Image')
plt.axis('off')

plt.subplot(3, 3, 7)
plt.axis('off')
plt.subplot(3, 3, 8)
plt.axis('off')
plt.subplot(3, 3, 9)
plt.axis('off')

plt.tight_layout()
plt.show()
```
