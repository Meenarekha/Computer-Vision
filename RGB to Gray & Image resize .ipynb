```python
#Using Libraries (OpenCV)
import cv2
import matplotlib.pyplot as plt
in_img=cv2.imread(r"C:\Users\student\Pictures\Lenna_(test_image).png")
in_img_rgb = cv2.cvtColor(in_img, cv2.COLOR_BGR2RGB)
gray_image= cv2.cvtColor(in_img, cv2.COLOR_BGR2GRAY)

fig=plt.figure()
imag1,imag2,imag3=fig.add_subplot(131),fig.add_subplot(132),fig.add_subplot(133)
imag1.imshow(in_img)
imag1.set_title('BGR Iamge')

imag2.imshow(cv2.cvtColor(in_img, cv2.COLOR_BGR2RGB))  
imag2.set_title('Original(RGB) Image')

imag3.imshow(gray_image, cmap='gray')
imag3.set_title('Grayscale Image')

plt.show()
```

```python
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
histogram
```

```python
plt.plot(histogram, color='black')
plt.xlim([0, 256])  # Set x-axis limits
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
```

```python
in_img=cv2.imread(r"C:\Users\student\Pictures\Lenna_(test_image).png")
new_width=250
new_height=150
resize_img=cv2.resize(in_img,(new_width,new_height))
plt.figure(figsize=(10,7))
plt.subplot(121)
plt.imshow(cv2.cvtColor(in_img,cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(resize_img,cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.show()
```

```python
in_img=cv2.imread(r"C:\Users\student\Pictures\Lenna_(test_image).png")
# Adjust contrast
alpha = 1.7 # Contrast control 
beta = 3 # Brightness control 
adjusted_img=cv2.convertScaleAbs(in_img,alpha=alpha,beta=beta)
plt.figure(figsize=(10,7))
plt.subplot(121)
plt.imshow(cv2.cvtColor(in_img,cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(adjusted_img,cv2.COLOR_BGR2RGB))
plt.title('Contrast Adjusted Image')
plt.show()
```
