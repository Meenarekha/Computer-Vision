```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
image1=cv2.imread(r"C:\Users\student\Pictures\cv\butterfly.jfif")

rgb_image = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

transformation_matrix = np.array([[0.299, 0.587, 0.114],
                                   [0.596, -0.274, -0.321],
                                   [0.211, -0.523, 0.311]])
reshaped_rgb = rgb_image.reshape((-1, 3))
yiq_image=np.dot(reshaped_rgb,transformation_matrix)
yiq_image = yiq_image.reshape(rgb_image.shape)
y_channel, i_channel, q_channel = cv2.split(yiq_image)


plt.figure(figsize=(15, 5))


plt.subplot(1, 4, 1)
plt.title('YIQ Image')
plt.imshow(yiq_image)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(y_channel)  
plt.title('Y Channel')
plt.axis('off')


plt.subplot(1, 4, 3)
plt.imshow(i_channel)  
plt.title('I Channel')
plt.axis('off')


plt.subplot(1, 4, 4)
plt.imshow(q_channel) 
plt.title('Q Channel')
plt.axis('off')


plt.show()
```
